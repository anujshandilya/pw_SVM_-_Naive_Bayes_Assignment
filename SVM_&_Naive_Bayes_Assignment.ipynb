# **Theory**

1. **What is a Support Vector Machine (SVM)?**
- A Support Vector Machine (SVM) is a powerful supervised learning algorithm used for classification and regression tasks. It works by finding the optimal decision boundary that separates data into different classes.
- Some features:
  - Effective in high-dimensional spaces.
  - Works well for both linear & non-linear classification.
  - Robust to overfitting, especially with proper kernel choice.

2. **What is the difference between Hard Margin and Soft Margin SYM?**
- Hard Margin SVM
  - Strict separation >> Assumes data is perfectly separable by a hyperplane.
  - No misclassification allowed >> The hyperplane must perfectly classify all training samples.
  - Maximizes the margin without tolerance for errors.
  - Best for: Linearly separable datasets without noise.
  - Problem: Fails on real-world noisy data where perfect separation isn’t possible.
- Soft Margin SVM
  - Allows some misclassification >> Introduces a slack variable ((\xi)) to handle noisy data.
  - Balances margin width & classification errors >> Finds a trade-off between maximizing the margin and allowing some errors.
  - Works well on real-world datasets with overlapping classes.
  - Best for: Real-world noisy datasets or cases with overlapping classes.
  - Problem: Over-relaxing constraints can reduce model robustness.

3. **What is the mathematical intuition behind SVM?**
- Support Vector Machines (SVM) have a strong mathematical foundation based on optimization and geometry. The core idea is to find the optimal hyperplane that maximizes the margin between different classes.

  - Finding the Optimal Hyperplane >> Given a binary classification problem, SVM seeks a hyperplane defined by:
  [ w \cdot x + b = 0 ]
    - where:
    - ( w ) → Weight vector (defines orientation of the hyperplane)
    - ( x ) → Feature vector (data points)
    - ( b ) → Bias (determines offset from origin)
  - Maximum Margin Principle >> SVM aims to maximize the margin, which is the distance between the hyperplane and the closest data points (support vectors). A larger margin means better generalization.

  - Optimization Using Lagrange Multipliers >> To maximize the margin, SVM solves:
  [ \min_{w, b} \frac{1}{2} ||w||^2 ]
  subject to: [ y_i (w \cdot x_i + b) \geq 1 ]
  where:
  - ( y_i ) → Class label (+1 or -1)
  - ( x_i ) → Feature vector
  This quadratic optimization ensures maximum separation between classes.

  - Soft Margin (Handling Misclassifications) >> For non-separable data, SVM introduces slack variables ((\xi)):
  [ y_i (w \cdot x_i + b) \geq 1 - \xi_i ]
  This allows some errors while still maximizing the margin.
  - Kernel Trick for Non-Linear Separation >> If classes are not linearly separable, SVM transforms data into a higher-dimensional space using kernel functions. Common kernels include:
    - Linear Kernel → (\langle x, x' \rangle)
    - Polynomial Kernel → ((x \cdot x' + c){-\gamma ||x - x'||^2} ) This allows SVM to separate data even if the decision boundary is curved.

4. **What is the role of Lagrange Multipliers in SVM?**
- Lagrange Multipliers play a crucial role in Support Vector Machines (SVMs) by helping solve the optimization problem that finds the best hyperplane for classification.
- They enable SVM to handle constraints mathematically and efficiently.

5. **What are Support Vectors in SVM?**
- Support Vectors in Support Vector Machines (SVM) are the critical data points that define the decision boundary.
- These points are closest to the hyperplane and play a key role in maximizing the margin between different classes.

6. **What is a Support Vector Classifier (SVC)?**
- A Support Vector Classifier (SVC) is a machine learning model that uses Support Vector Machines (SVMs) for classification tasks.
- It finds the optimal hyperplane that separates different classes in a dataset while maximizing the margin between them.

7. **What is a Support Vector Regressor (SVR)?**
- A Support Vector Regressor (SVR) is a machine learning algorithm that applies Support Vector Machines (SVMs) to regression problems instead of classification.
- SVR works by finding a function that best predicts a continuous target while maximizing the margin and ignoring small errors.

8. **What is the Kernel Trick in SVM?**
- The Kernel Trick in Support Vector Machines (SVMs) is a powerful technique that allows SVM to handle non-linearly separable data by mapping it into a higher-dimensional space, where it becomes linearly separable.
- Some Commonly used kernels are:
  - Linear kernel
  - Poynomial kernel
  - Rbf(Radial Basis Function) kernel
  - Sigmoid kernel

9. **Compare Linear Kernel, Polynomial Kernel, and RBF Kernel.**
- The Linear Kernel, Polynomial Kernel, and Radial Basis Function (RBF) Kernel are widely used in Support Vector Machines (SVMs) for different types of classification and regression tasks. Each has its strengths and weaknesses, depending on the nature of the data.
- Linear Kernel
  - Best for linearly separable data.
  - Computationally efficient, especially for high-dimensional datasets like text classification.
  - Example applications: Spam detection, document classification.
  - Advantage:
    - Fast computation.
    - Works well when data follows a linear decision boundary.
    - Used in high-dimensional spaces (like text data).
  - Disadvantage:
    - Fails on non-linear data (requires feature engineering).
- Polynomial Kernel
  - Best when relationships between features involve higher-order interactions.
  - Captures curved decision boundaries without explicit feature engineering.
  - Example applications: Handwriting recognition, image classification.
  - Advantage:
    - Can model complex relationships between features.
    - Suitable for moderate-sized datasets with curved boundaries.
  - Disadvantage:
    - Computationally expensive for large datasets.
    - Choosing the right degree ( d ) can be tricky—too high may cause overfitting
- Radial Basis Function (RBF) Kernel
  - Best for highly non-linear data.
  - Maps data into infinite-dimensional space, allowing complex separation.
  - Example applications: Bioinformatics, facial recognition.
  - Advantage:
    - Powerful for complex decision boundaries (works for almost any dataset).
    - Less dependent on feature selection compared to other kernels.
  - Disadvantage:
    - Requires careful tuning of ( \gamma ) parameter—too high may cause overfitting.
    - Computationally expensive on large datasets.


10. **What is the effect of the C parameter in SVM?**
- The C parameter in Support Vector Machines (SVMs) controls the trade-off between maximizing the margin and minimizing classification errors.
- It influences how much the model penalizes misclassifications
- Effect of C:
  - Higher C (Strict Classification, Low Margin) >> The model tries to classify all points correctly and allows a smaller margin.
  - Lower C (Flexible Classification, Larger Margin) >> The model prioritizes a wider margin and tolerates some misclassifications to improve generalization


11. **What is the role of the Gamma parameter in RBF Kernel SVM?**
- The Gamma ((\gamma)) parameter in Radial Basis Function (RBF) Kernel SVM controls the influence of individual data points on the decision boundary.
- It affects how the model maps data into a higher-dimensional space for classification
- Effect of Gamma:
  - High Gamma ((\gamma)):
    - Tight Decision Boundaries
    - Points have a strong influence on classification.
    - More complex decision boundaries, leading to potential overfitting.
  - Low Gamma ((\gamma)):
    - Loose Decision Boundaries
    - Points have a weaker influence on classification.
    - Simpler decision boundary, improving generalization but potentially underfitting

- It is called Naive because:
  - The term "Naive" comes from the assumption that features are independent given the class, meaning:

13. **What is Bayes’ Theorem?**
- Bayes’ Theorem is a fundamental principle in probability theory and statistics that describes how to update the probability of an event based on new evidence. It plays a key role in machine learning, especially in Naive Bayes classifiers and Bayesian inference.
- It is given by:
  - p(A/B) = p(A)*p(B/A)/P(B)
    - Where
      - p(A): Probability of event A.
      - p(B): Probability of event B.
      - p(A/B): Probability of event A given B.
      - p(B/A): Probability of event B given A.

14. **Explain the differences between Gaussian Naive Bayes, Multinomial Naive Bayes, and Bernoulli Naive Bayes.**
- The Gaussian Naive Bayes (GNB), Multinomial Naive Bayes (MNB), and Bernoulli Naive Bayes (BNB) classifiers are different variants of Naive Bayes, each tailored for specific types of data distributions. Here’s a breakdown of their differences:
  - Gaussian Naive Bayes (GNB)
    - Best for: Continuous numerical data
    - Assumption: Features follow a normal (Gaussian) distribution
    - Applications:
      - Medical diagnosis (e.g., predicting diseases based on continuous health metrics)
      - Stock market prediction (e.g., analyzing price fluctuations)
    - Limitation: Works only with continuous features—categorical features don’t fit well.
  - Multinomial Naive Bayes (MNB)
    - Best for: Text classification problems
    - Assumption: Features represent count-based frequency (e.g., word occurrences in documents)
    - Applications:
      - Spam filtering (e.g., detecting spam based on word distribution)
      - Sentiment analysis (e.g., analyzing words to classify positive/negative reviews)
    - Limitation: Works best with count data (like word frequencies)—not suitable for binary or continuous features.
  - Bernoulli Naive Bayes (BNB)
    - Best for: Binary data (0 or 1 values)
    - Assumption: Features are binary (present or absent, yes or no)
    - Applications:
      - Text classification (e.g., spam detection, sentiment analysis using binary word indicators)
      - Fraud detection (e.g., analyzing transaction patterns as fraud/not fraud)
    - Limitation: Works best when features are truly binary—not ideal for word frequencies or continuous values.

15. **When should you use Gaussian Naive Bayes over other variants?**
- You should use Gaussian Naive Bayes (GNB) when your features are continuous and approximately follow a normal distribution.
-  Unlike Multinomial Naive Bayes (MNB) and Bernoulli Naive Bayes (BNB), which are optimized for categorical or binary data, GNB assumes Gaussian (normal) distribution for each feature, making it ideal for numerical datasets.

16. **What are the key assumptions made by Naive Bayes?**
- The Naive Bayes classifier makes several key assumptions to simplify probability calculations and efficiently classify data. These assumptions drive both the model’s strengths and limitations.
- Following are the key assumptions:
  - The model assumes that features are conditionally independent given the class label.
  - The model treats each feature as equally important, without weighting them differently
  - It assumes that the presence of a feature is independent of other features within each class.

17. **What are the advantages and disadvantages of Naive Bayes?**
- Advantages of Naive Bayes
  - Fast & Computationally Efficient → Works well on large datasets with minimal training time.
  - Handles High-Dimensional Data → Performs well in situations where there are many features, such as text classification.
  - Works Well Even with Limited Data → Requires less training data compared to complex models like neural networks.
  - Interpretable & Simple → Provides intuitive probabilities for classification decisions.
  - Resistant to Overfitting → Works surprisingly well even if feature independence assumption is violated.
  - Performs Well in Text Classification & NLP → Used for spam detection, sentiment analysis, and topic modeling.

- Disadvantages of Naive Bayes
  - Strong Independence Assumption → Assumes features are independent, which rarely holds in real-world data.
  - Feature Importance Not Considered → Unlike Decision Trees or Random Forests, it treats all features equally.
  - Doesn’t Handle Correlated Features Well → If features are strongly correlated, the model may struggle to capture interactions.
  - Continuous Features Require Gaussian Assumption → Works best with normally distributed continuous data (for Gaussian NB).

18. **Why is Naive Bayes a good choice for text classification?**
- Naive Bayes is one of the best choices for text classification, especially in applications like spam detection, sentiment analysis, and topic categorization because:
  - Text data has many features (e.g., thousands of words in documents).
  - Naive Bayes efficiently handles large vocabularies without performance issues.
  - Unlike models like SVM, it doesn’t require complex feature engineering.
  - Computationally efficient → It learns quickly, even with large datasets.
  - Real-time classification → Ideal for tasks like spam filtering where speed is crucial.
  - Text data is sparse (many words are absent in documents).
  - Multinomial Naive Bayes handles zero-frequency words well using Laplace smoothing.
  - Performs well even with small datasets → Unlike deep learning, which needs tons of data.
  - Doesn’t overfit easily → Works fine even if the training data is not perfectly balanced.
  - Naive Bayes calculates probabilities → Useful for confidence-based predictions.

19. **Compare SVM and Naive Bayes for classification tasks.**
- SVM:
  - Finds an optimal decision boundary for classification, works well with complex, continuous data, and is more accurate but computationally expensive. Sensitive to noisy data and best for small datasets.
  - Use SVM for: Precise boundaries, smaller datasets, continuous features.
- Naive Bayes:
  - Uses probabilistic inference with the assumption that features are independent. Ideal for text classification (spam detection, sentiment analysis), fast computation, and large datasets but may struggle with correlated features.
  - Use Naive Bayes for: Text classification, large datasets, fast predictions


20. **How does Laplace Smoothing help in Naive Bayes?**
- Laplace Smoothing helps Naive Bayes by addressing the zero-frequency problem, which occurs when a word or feature never appears in the training data for a particular class. Without smoothing, Naive Bayes assigns zero probability to such cases, leading to incorrect predictions.
- It is useful by:
  - Prevents zero probabilities, ensuring smooth classification.
  - Improves handling of rare words in text classification (spam filtering, sentiment analysis).
  - Enhances model generalization, especially for small datasets.

21. Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy.

```python
import pandas as pd

#Load the Dataset iris
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['target'] = iris.target

#Seperate x and y
x = df.drop('target', axis = 1)
y = df['target']

# Train-test-split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1)

# Model training
from sklearn.svm import SVC
svc = SVC(kernel='linear')
svc.fit(x_train, y_train)

# Model Prediction
y_pred = svc.predict(x_test)

# Evaluation
from sklearn.metrics import accuracy_score
print(f"The accuracy of the model is {accuracy_score(y_test, y_pred)}")
```

24. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision
boundary.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Load iris dataset and use only two classes and two features
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data[:, :2]  # Use only first two features
y = iris.target

# Filter to two classes (e.g., class 0 and 1)
X = X[y != 2]
y = y[y != 2]

from sklearn.svm import SVC
model = SVC(kernel='linear', C=1.0)
model.fit(X, y)

# Create a mesh grid over the feature space
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

# Predict over the mesh
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision surface
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)

# Plot original data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')

# Plot support vectors
plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
            s=100, facecolors='none', edgecolors='black', label='Support Vectors')

# Labels and title
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('SVM Decision Surface')
plt.legend()
plt.show()
```

26. Write a Python program to train a Multinomial Naive Bayes classifier for text classification using the 20
Newsgroups dataset.

```python
import pandas as pd

# Load the Dataset
from sklearn.datasets import fetch_20newsgroups
categories = ['sci.space', 'rec.sport.baseball', 'comp.graphics', 'talk.politics.mideast']
data = fetch_20newsgroups(subset='train', categories=categories, remove=('headers', 'footers', 'quotes'))
test_data = fetch_20newsgroups(subset='test', categories=categories, remove=('headers', 'footers', 'quotes'))
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')
x_train = vectorizer.fit_transform(data.data)
x_test = vectorizer.transform(test_data.data)
y_train = data.target

# Training of Dataset
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train, data.target)

y_pred = model.predict(x_test)

# Evaluation
from sklearn.metrics import accuracy_score, classification_report
accuracy = accuracy_score(test_data.target, y_pred)
print(f"Accuracy: {accuracy}")
print("Classification Report:\n", classification_report(test_data.target, y_pred, target_names=categories))
```

33. Write a Python program to implement a Naive Bayes classifier for spam detection using email data.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Load SMS Spam dataset (Ensure dataset exists or use another email dataset)
url = "https://raw.githubusercontent.com/Apaulgithub/oibsip_taskno4/main/spam.csv"
df = pd.read_csv(url, encoding='latin-1')
df.head()
df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis = 1, inplace = True)

# Rename columns
df.columns = ['label', 'message']

# Convert labels to binary (spam = 1, ham = 0)
df['label'] = df['label'].map({'spam': 1, 'ham': 0})

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df['message'], df['label'], test_size=0.2, random_state=42)

# Convert text into TF-IDF feature vectors
vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Train Naive Bayes spam classifier
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)

# Predict on test set
y_pred = model.predict(X_test_tfidf)

# Evaluate model performance
accuracy = accuracy_score(y_test, y_pred)
print(f"Spam Detection Accuracy: {accuracy}")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
